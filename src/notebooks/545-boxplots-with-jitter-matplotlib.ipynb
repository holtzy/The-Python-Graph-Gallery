## Libraries

First, you need to install the following librairies:
- [matplotlib](https://python-graph-gallery.com/matplotlib/) is used for creating the plot
- `pandas` for data manipulation
- `numpy` for data generation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Dataset

We use `np.random.normal()` to create a **normally distributed** variable and `pandas` to put it into a dataframe.

```python
# Generate data for Group A (Normal Distribution)
value = np.random.normal(loc=50, scale=10, size=100)# Create a DataFrame
df = pd.DataFrame({'value': value})
```

## Basic boxplot with jitter

Once we've opened our dataset, we'll now **create the graph**. The following displays the **distribution** using the `boxplot()` function.

Adding **individual data points**  with [matplotlib](https://python-graph-gallery.com/matplotlib/) is actually adding a [scatter plot](https://python-graph-gallery.com/scatter-plot/) on top of the main chart. 

We need to specify the **maximum width** over which the points will be represented, and then generate their positions randomly, ideally **uniformly** (`np.random.uniform()`) for best visibility. However, this is perfectly **customizable**. 

```python
# Create a boxplot with jitter
fig, ax = plt.subplots()

# Create boxplot
ax.boxplot(df['value'])

# Add jittered points
lower_born = 0.92
higher_born = 1.08
positions = np.random.uniform(lower_born, higher_born,
                              size=len(df))
ax.scatter(positions,
           df['value'])

# Add a title
ax.set_title('Boxplot with Jitter')

# Show the plot
plt.show()
```

## Custom markers

To make the graph more readable, we need to change several properties of our boxplots.

- size of the markers: `s` argument
- style of the markers: `marker` argument (must be in the following list: `o`, `s`, `d`, `^`, `v`, `<`, `>`, `p`, `*`, `+`, `x`, `h`, `1`, `2`, `3`, `4`)

```python
# Create a boxplot with jitter
fig, ax = plt.subplots()

# Create boxplot
ax.boxplot(df['value'])

# Add jittered points
lower_born = 0.92
higher_born = 1.08
positions = np.random.uniform(lower_born, higher_born,
                              size=len(df))
ax.scatter(positions,
           df['value'],
           s=50,
           marker='*',
           color='gold',
           edgecolor='black',
           linewidth=0.3)

# Add a title
ax.set_title('Boxplot with Jitter')

# Show the plot
plt.show()
```

## Flip the plot

In order to **flip the plot** on the other axis, we have to:
- add the `vert=False` argument in the `boxplot()` function
- switch the **first and the second** argument in the `scatter()` function

```python
# Create a boxplot with jitter
fig, ax = plt.subplots()

# Create boxplot
ax.boxplot(df['value'],
           vert=False)

# Add jittered points
lower_born = 0.92
higher_born = 1.08
positions = np.random.uniform(lower_born, higher_born,
                              size=len(df))
ax.scatter(df['value'],
           positions,
          )

# Add a title
ax.set_title('Boxplot with Jitter')

# Show the plot
plt.show()
```
