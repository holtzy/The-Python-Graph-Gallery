## Libraries

First, you need to **install** `drawarrow` with the following command: `pip install drawarrow`.

With `drawarrow` you can easily draw curved, straight, thin, large or bizarre arrows in [matplotlib](/matplotlib), in any color, width and style you like. It was created by [Joseph Barbier](https://barbierjoseph.com/).

`drawarrow` provides 2 functions: `fig_arrow()` (draw an arrow on a matplotlib figure) and `ax_arrow()` (draw an arrow on a matplotlib axes). We'll also need to load [matplotlib](/matplotlib) to create the figure.


```python
import matplotlib.pyplot as plt
from drawarrow import fig_arrow, ax_arrow
```

## Quick start

The easiest way to get started is to specify a **minimalist arrow** to highlight a point in a [scatter plot](/scatter-plot)


```python
fig, ax = plt.subplots()
ax.scatter(x=[1, 2, 3, 4, 5], y=[1, 2, 3, 4, 5], s=100)
fig_arrow(
   head_position=(0.5, 0.5),
   tail_position=(0.2, 0.7),
   fig=fig,
   color="red"
)
plt.show()
```

## Draw an arrow on an axes

Matplotlib graphs have **two main components**: the figure (the overall graph) and the axes (all the sub-graphs).

The `ax_arrow()` function is designed to **draw arrows on an axes** (for example, using the same coordinate system as the plotted data).


```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
ax1.scatter(x=[1, 5, 10], y=[5, 12, 3], s=300)
ax2.scatter(x=[1, 5, 10], y=[5, 12, 3], s=300)
ax_arrow(
   tail_position=(2, 8),
   head_position=(9.5, 3),
   ax=ax1,
   color="green",
   tail_width=10,
   head_width=30,
   head_length=30,
   radius=0.4
)
plt.show()
```

## Draw an arrow on a figure

The `fig_arrow()` function is designed to **draw arrows on a figure** (using the relative coordinate system).


```python
fig, ax = plt.subplots()
ax.scatter(x=[1, 2, 3, 4, 5], y=[1, 2, 3, 4, 5], s=300)
fig_arrow(
   head_position=(0.48, 0.5),
   tail_position=(0.2, 0.7),
   fig=fig,
   tail_width=5,
   head_width=20,
   head_length=20,
)
plt.show()
```

## Custom style

As the arrow displayed is in fact a `FancyArrowPatch`, you can pass all its additional arguments to `fig_arrow()` and `ax_arrow()`. The most useful of these are used in the example below.


```python
fig, ax = plt.subplots()
ax.scatter(x=[1, 2, 3, 4, 5], y=[1, 2, 3.2, 4, 5], s=300)
fig_arrow(
   head_position=(0.48, 0.5),
   tail_position=(0.2, 0.7),
   fig=fig,
   facecolor="#003049",
   edgecolor="#d62828",
   linewidth=1,
   radius=1.2,
   alpha=0.9,
   hatch="O",
   tail_width=5,
   head_width=20,
   head_length=20,
)
plt.show()
```
