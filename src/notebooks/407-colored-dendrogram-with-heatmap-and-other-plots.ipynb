This example draws a heatmap with dendrograms colored by clusters and a line plot of mean values of rows.  

```python
import seaborn as sns
from omniplot import plot as op
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat=np.concatenate([0.25*np.random.uniform(0,1,size=[10, 25]),0.5*np.random.uniform(0,1,size=[15, 25]),
np.random.uniform(0,1,size=[15, 25])])
mat=np.concatenate([mat, np.random.uniform(0,1,size=[40, 15])], axis=1)
df=pd.DataFrame(mat)
```

```python
res=op.heatmap(df,
               row_plot={"mean": mat.mean(axis=1)} ,
               cbar_title="random colors",
               approx_clusternum=3,
               ztranform=False)
```

And, it can change the elements of the heatmap by adding the sizes option.

```python

res=op.heatmap(df,
               sizes=df.values,
               row_plot={"mean": mat.mean(axis=1)} ,
               cbar_title="random colors",
               approx_clusternum=3,
               ztranform=False)
```
